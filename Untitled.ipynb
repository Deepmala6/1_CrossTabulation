```python
# Cross-tabulation 1: Count of Sales Channel by Region
crosstab_1 = pd.crosstab(
    index=df["Region"],
    columns=df["Sales_Channel"]
)

# Display the result
print("Crosstab 1: Sales Channel by Region\n")
print(crosstab_1)
```

```python
# Cross-tabulation 2: Count of Order Priority by Sales Channel
crosstab_2 = pd.crosstab(
    index=df["Sales_Channel"],
    columns=df["Order_Priority"]
)

# Display the result
print("\nCrosstab 2: Order Priority by Sales Channel\n")
print(crosstab_2)
```

```python
# Cross-tabulation 4: Total Revenue by Country and Sales Channel
crosstab_3 = pd.crosstab(
    index=df["Country"],
    columns=df["Sales_Channel"],
    values=df["Total_Revenue"],
    aggfunc="sum"
)

# Display the result
print("\nCrosstab 3: Total Revenue by Country and Sales Channel\n")
print(crosstab_3)
```

```python
# Cross-tabulation 5: Average Unit Cost by Item Type and Region
crosstab_4 = pd.crosstab(
    index=df["Item_Type"],
    columns=df["Region"],
    values=df["Unit_Cost"],
    aggfunc="mean"
)

# Display the result
print("\nCrosstab 4: Average Unit Cost by Item Type and Region\n")
print(crosstab_4)
```
